# Preprocessing

# 1. Case Folding

```python
def casefolding(data):
    data = data.lower()
    return data


review = []
for index, row in dataset.iterrows():
    review.append(casefolding(row["Review"]))
    
dataset["Review"] = review
```

```python
dataset.head()
```

# 2. Cleaning

```python
import string, re
PUNCT_TO_REMOVE = string.punctuation
def cleaning(text):
    # lower text
    
    
 
    text = re.sub(r'''     
               \W+     
               \s*     
               ''',    
               ' ',    
               text,
               flags=re.VERBOSE)
    text = re.sub(r"\d+", " ", text)
    text = text.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ")
    text = text.encode('ascii', 'replace').decode('ascii')
    text = ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ", text).split())
    text = text.translate(str.maketrans(' ', ' ', PUNCT_TO_REMOVE))
    text = text.replace("http://", " ").replace("https://", " ")
    text = ' '.join(re.sub(r"\d+", "", text).split())
    
    return re.sub(r"\b[a-zA-Z]\b", "", text)

review = []
for index, row in dataset.iterrows():
    review.append(cleaning(row["Review"]))
    
dataset["Review"] = review
```

```python
dataset.head()
```

# 3. Tokenization

```python
dataset['Review'] = [word_tokenize(word) for word in dataset["Review"] ]
```

```python
dataset.head()
```

# 4. Word Normalization

```python

kamus_normalisasi = {}

for index, row in kamus.iterrows():
    if row[0] not in kamus_normalisasi:
        kamus_normalisasi[row[0]] = row[1]

def normalisasi_kata(text):
    return [kamus_normalisasi[word] if word in kamus_normalisasi else word for word in text]
dataset['Review'] = dataset['Review'].apply(lambda text: normalisasi_kata(text))
```

```python
dataset.head()
```

# 5. Stopword Removal

```python
from nltk.corpus import stopwords
STOPWORDS = stopwords.words('indonesian')
# print(STOPWORDS)
STOPWORDS.extend(['amp','hahaha','ya','bnr','bener','iya','dikit','sih','nya', 'yg', 'dll','gara', 'dan lain', 'dan lain-lain', 'oh','y','g','yes', 'yang','gak','bgt','ku','buat','karena','karna','ga','udah','udh','pas','kalau','kalo','eh','deh','coba','oke','a','aa','aaa','aaaa','aaaaa','huhuhu','hihihi','hahaha','ha','haha','hehehe','he','heh','hehe','si','ah','aja','ni','ini','xixi','da','woy'])

stopword_list = set(STOPWORDS)
def stopword_removal(text):
    return [word for word in text if word not in stopword_list]

dataset['Review'] = dataset['Review'].apply(lambda text: stopword_removal(text))
dataset.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = TfidfVectorizer(ngram_range=(3,3)).fit(dataset["Review"].apply(lambda tokens: ' '.join(tokens)))
vectorizer.get_feature_names_out()
for i in vectorizer.get_feature_names_out():
    print(i)
```

```python

from sklearn.model_selection import train_test_split
X = dataset['Review'].apply(lambda tokens: ' '.join(tokens))
y = dataset["Aroma"]
```

# **SPLIT DATA LATIH DAN DATA UJI**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.4,
                                                    stratify=y,
                                                    random_state=42)
```

```python
X_train_vector = vectorizer.transform(X_train).toarray()
X_test_vector = vectorizer.transform(X_test).toarray()

# tfidf = pd.DataFrame(X_train_vector,
#              index = X_train.index,
#              columns = vectorizer.get_feature_names_out()).T

# tfidf
```
